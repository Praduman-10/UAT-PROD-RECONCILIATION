```python
def is_numeric(s):
    try:
        # Try to convert the string to a float
        float(s)
        return True
    except ValueError:
        return False


def highlight_cells(val):
    if is_numeric(val):
        if not isinstance(val, float):  # Check if it's not a float
            color = 'background-color: lavender'
        else:
            color = ''
    else:
        if not isinstance(val, float):  # Check if it's not a float
            color = 'background-color:lightblue'
        else:
            color = ''
    font_weight = 'font-weight: bold'
    return f'{color}; {font_weight}'

# Apply the highlight function to the DataFrame
styled_df = final_valid.style.applymap(highlight_cells)
```
